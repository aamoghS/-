Concat() function: Combines two or more dataframes


```python
# Create a DataFrame df1 with columns 'A' and 'B', containing sample data
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 'B': ['B0', 'B1', 'B2', 'B3']})

# Display the label 'df1:' along with the contents of the df1 DataFrame
display('df1:', df1)

# Create a DataFrame df2 with columns 'A' and 'B', containing sample data
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 'B': ['B4', 'B5', 'B6', 'B7']})

# Display the label 'df2:' along with the contents of the df2 DataFrame
display('df2:', df2)
```

```python
display(pd.concat([df1, df2])) # combines the dataframes together into one dataframe
```

transform() function: Modifies each value of a dataframe column using a given function

```python
df = pd.DataFrame({'A': [0.0, 1.5, 2.3, 3.9], 'B': [0, 1, 2, 3]})
display('df:', df)
print(df.info())
```

```python
df['A'] = df['A'].transform(int) # truncates the A values from 1.5 to 1
display(df)
print(df.info())
```

```python
df['A'] = df['A'].transform(lambda x : x + 5) # adds 5 to each
display(df)
```

```python
df['AB'] = df['A'] + df['B'] #adds the A and B values together
display(df)
```
